```python
import pandas as pd
```

```python
df = pd.read_csv('/content/salaries.csv')
df
```

```python
inputs = df.drop(['salary_more_then_100k'],axis ='columns' )
target = df['salary_more_then_100k']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
LabelEncoder()
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis = 'columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,2,1]])
```

```python
model.predict([[2,0,1]])
```

```python
model.predict([[2,0,2]])
```

```python
pk = pd.read_csv('/content/titanic.csv')
pk
```

```python
pk.info()
```

```python
pk.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis = 'columns')
```

```python
inputs = pk.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns')
target = pk['Survived']
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs.Age[:66]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,target,test_size = 0.2)
```

```python
x_train.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=pk)
plt.title('Overall Survival Rate')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```
