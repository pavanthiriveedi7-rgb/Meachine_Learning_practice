```python
import pandas as pd
df = pd.read_csv('/content/homeprices (1).csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
dummies['monroe township'] = dummies['monroe township'].astype(int)
dummies['west windsor'] = dummies['west windsor'].astype(int)
dummies['robinsville'] = dummies['robinsville'].astype(int)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns')

merged
```

```python
final = merged.drop(['town','west windsor'],axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop(['price'], axis='columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[2000,0,1]])
```
