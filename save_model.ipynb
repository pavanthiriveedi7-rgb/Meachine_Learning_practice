```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/homeprices.csv')
```

```python
p = linear_model.LinearRegression()
p.fit(df[['area']],df['price'])
```

```python
p.predict(np.array([[5000]]))
```

```python
p.predict([[5000]])
```

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
  pickle.dump(p,f)
```

```python
with open('model_pickle','rb') as f:
  mp = pickle.load(f)
```

```python
mp.predict([[5000]])
```

```python
import joblib
```

```python
joblib.dump(p,'model_joblib')
```

```python
mj = joblib.load('model_joblib')
```

```python
mj.predict([[5000]])
```

```python
mj.coef_
```

```python
mj.intercept_
```

### 1. Efficiency with Large Data
The biggest reason to choose one over the other is NumPy.

Pickle treats a massive NumPy array as a standard Python object. This can lead to slow "copying" and high memory usage.

Joblib detects NumPy arrays and handles them using specialized "fast-path" code. Instead of loading everything into memory at once, it uses memory mapping (mmap), which makes reading large datasets much faster.

### 2. The Multi-File Quirk
When you save a complex object with Pickle, you get one single file. When you save a model with Joblib (especially older versions), it might create one main file and several sub-files (like .npy files) for each large array in your model. If you move your model to another folder, you have to remember to move all those files, not just the main one.

### 3. Security Warning
Important Note: Both Pickle and Joblib are not secure. Never "unpickle" a file you downloaded from an untrusted source. It is possible to craft a malicious file that executes arbitrary code on your machine the moment you try to load it.

### Which one should you use?
Use Pickle for small, standard Python objects (like a list of user settings or a dictionary of strings).

Use Joblib for saving and loading Scikit-learn models, large NumPy arrays, or anything involving heavy numerical data.
